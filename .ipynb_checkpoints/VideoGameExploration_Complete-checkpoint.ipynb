A bar graph would be a better way to visualize how many games in our dataset were released each year.

#### Goal: Create a bar graph showing the number of games in our dataset released each year.

Programming requires breaking a goal into smaller steps. The steps we'll take here are:

1. Isolate the 'Year' column of the dataframe.
2. Count how many games appear in each year.
3. Sort those counts by year.
4. Create a bar chart of that information.

We *could* do these four steps in one line of code, but it's often more clear to split a long, complicated line of code into multiple lines of code. A code cell in a Python notebook can contain any number of lines of code. To keep it even clearer, we can insert *comments* in-line with the code. Comments are human readable lines that help programmers understand what is going on; they don't get executed as code. In the cell below, notice that all the comments start with the # syntax.

Between each step, we'll store the result in a variable. Again, we can call our variables by any name, but we'll try to choose names that make the expected contents clear. Then, when we use those name in future steps, we know that we are referring to the output from previous steps.

**Read through each step of the code cell below, noting the different methods and syntax used. Then, run the cell to see the bar graph.**

```python
%matplotlib inline
```

```python
# 1. Isolate the 'Year' column of the dataframe.
year_column = df['Year']

# 2. Count how many games appear in each year.
year_counts = year_column.value_counts()

# 3. Sort those counts by year.
sorted_years = year_counts.sort_index()

# 4. Create a bar chart of that information.
sorted_years.plot.bar()
```

Using substitution for the variables above, we could rewrite the final line, **sorted_years.plot.bar()**

instead as

**df['Year'].value_counts().sort_index().plot.bar()**

It's up to us as programmers to balance brevity and clarity.

Now that we have a bar graph, we can visualize the release year distribution much more easily.
